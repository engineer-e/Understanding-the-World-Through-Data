# Questions on Visualizations

In this series of questions you will be deciding which datasets match each plot. Create the following cleaned up dataset called `clean`, which is the full temperature dataset without any -99 temperatures. Use the command below (run the cell below), assuming that your full dataset is saved in a DataFrame called df2.

```python
clean = df2.loc[(df2["AvgTemperature"] != -99)]
```

## Question 1

Create an x-y plot of the average temperature in Paris in the year 2002.

```python
# Your code here
import matplotlib.pyplot as plt
average_temperature = clean[(clean["Year"]==2002) & (clean["City"]=="Paris")]["AvgTemperature"]

plt.plot(range(len(average_temperature)), average_temperature)
plt.title('Average Temperature in Paris (2002)')
plt.xlabel('Day of Year')
plt.ylabel('Average Temperature')
plt.grid(True)
plt.show()
```

## Question 2

Create a scatter plot of the average temperature in Tokyo in the year 1996.

```python
# Your code here
import matplotlib.pyplot as plt
average_temperature = clean[(clean["Year"]==1996) & (clean["City"]=="Tokyo")]["AvgTemperature"]
plt.scatter(range(len(average_temperature)), average_temperature)
plt.title('Average Temperature in Tokyo (1996)')
plt.xlabel('Day of Year')
plt.ylabel('Average Temperature')
plt.grid(True)
plt.show()
```

## Question 3

Create a histogram of the average temperature in the country Sweden in March with 10 bins.

```python
# Your code here
import matplotlib.pyplot as plt
average_temperature = clean[(clean["Month"]==3) & (clean["Country"]=="Sweden")]["AvgTemperature"]

bins = average_temperature.count()//80
plt.hist(average_temperature, bins=bins)
plt.title('Average Temperature in Sweden in March')
plt.xlabel('Average Temperature')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Question 5

What is the correlation between average temperature and month for the country Japan in the year 2000 and months January through July?

```python
import seaborn as sns

correlation = clean[(clean["Year"]==2000) & (clean["Country"]=="Japan") & ((clean["Month"]>=1) & (clean["Month"]<=7))]
corr=correlation.corr(numeric_only=True)
_= sns.heatmap(corr)

# Explicitly print the correlation between AvgTemperature and Month
print(f"Correlation between Average Temperature and Month: {corr['AvgTemperature']['Month']:.2f}")
```
